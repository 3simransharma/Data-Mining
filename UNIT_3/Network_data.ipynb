```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import networkx as nx
```

```python
#Creating an an undirected graph
G = nx.Graph()
```

```python
# Adding Nodes (People, entity, etc..)
G.add_nodes_from(["Alice", "Bob", "Charlie", "David", "Eva"])

#Nodes are stored internally as graph vertices
```

```python
# Adding Edges (Relationships between the Nodes)

G.add_edges_from([
    ("Alice", "Bob"),
    ("Alice", "Charlie"),
    ("Bob", "David"),
    ("Charlie", "David"),
    ("David", "Eva")
])
```

```python
#Node-Link Diagram Using Default Layout

plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_size=1000)
plt.title("Node-Link Diagram (Default Layout)")
plt.show()

# -- nx.draw() draws nodes and edges
# -- with_labels=True means it shows names on nodes
# -- node_size=1000 means it controls node size
# -- Default layout places nodes automatically
```

```python
# Experiment with Layout Algorithms (Spring Layout)
pos = nx.spring_layout(G)

plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_size=1000)
plt.title("Spring Layout Visualization")
plt.show()
```

#Spring layout
Spring layout is a force-directed layout algorithm that positions nodes in a network by simulating attractive forces between connected nodes and repulsive forces between all nodes to produce a readable graph structure.

```python
# Apply Visual Encoding (Node Size & Color by Degree)
degrees = dict(G.degree())

# Degree = number of connections a node has
# Stored as a dictionary
# Example: {Alice: 2, David: 3}
```

```python
node_sizes = [degrees[node] * 500 for node in G.nodes()]
node_colors = [degrees[node] for node in G.nodes()]

# -- Nodes with higher degree means bigger size
# -- Node color intensity represents connectivity
```

```python
# Draw Graph with Visual Encoding

nx.draw(
    G, pos,
    with_labels=True,
    node_size=[1000,2000,1000,1000,1000],
    node_color=['red','blue','pink','green','yellow'],
    cmap=plt.cm.Blues
)
plt.title("Node-Link Diagram with Visual Encoding")
plt.show()

# node_size: shows importance
# node_color: shows connectivity
# cmap: color gradient
```

```python
# Create Adjacency Matrix Representation
adj_matrix = nx.adjacency_matrix(G).todense()
print("Adjacency Matrix:")
adj_matrix

# Converts network into matrix form
# Rows and columns represent nodes
# 1 means connection exists
# 0 means no connection
```

```python
# Display Matrix with Labels

adj_df = pd.DataFrame(adj_matrix, index=G.nodes(), columns=G.nodes())
print(adj_df)
```
