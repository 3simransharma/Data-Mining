The **StandardScaler** transforms data such that each feature has:

    A mean of 0
    A standard deviation of 1

This process is called standardization (or Z-score normalization). Unlike simple rescaling, it changes the distribution of the feature so that values are measured in terms of their distance (in standard deviations) from the mean. This is particularly useful when:

    The dataset follows (or approximately follows) a normal distribution.
    Algorithms assume standardized input (e.g., Linear Regression, Logistic Regression, PCA).

```python
#Standardization on Multiple columns in one go
df1 = df.copy()
cols = ["Age", "Fare", "SibSp", "Parch"]

scaler = StandardScaler()
df1[cols] = scaler.fit_transform(df1[cols])
df1
```

```python
#Manual Standardization
df2 = df.copy()
df2["Age_std"] = (df2["Age"] - df2["Age"].mean()) / df2["Age"].std()
df2
```

<h1>Normalization</h1>
Normalization is a scaling technique that transforms values into a fixed range, usually 0 to 1.

Both **Normalization and Standardization** have an aim to bring all values of different attributes to a similar scale, they are differently made because different algorithms work best with different types of scaling.

* Some algorithms work best with 0–1 range = Normalization
* Some algorithms work best with mean 0, std. 1 = Standardization

We can use **Normalization** for algorithms such as, **KNN, K-Means, Neural networks** because Normalization works best in these algorithms.

We can use **Standardization** for algorithms such as, **Logistic Regression, SVM, Linear Regression** because Standardization works best in these algorithms.

<H1>Difference Between <b> Normalization and Standardization</b></H1>

| **Normalization**                                      | **Standardization**                                 |
|--------------------------------------------------------|------------------------------------------------------|
| Scales data to a fixed range (usually 0 to 1)          | Centers data around mean 0 and standard deviation 1 |
| **Formula:** X' = (X − X_min) / (X_max − X_min)        | **Formula:** Z = (X − μ) / σ                        |
| Always between 0 and 1                                  | Values can be positive or negative                  |
| Best for distance-based algorithms (KNN, K-Means)       | Best for models needing normal distribution (SVM, PCA) |
| Used in K-Means, KNN, Neural Networks                   | Used in Logistic Regression, SVM, PCA, Linear Regression |
|  Sensitive to outliers                                | Less sensitive                                    |
