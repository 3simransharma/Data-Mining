```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import plotly.express as px
```

```python
customers = pd.read_csv("customers(1).csv")
purchases = pd.read_csv("purchases.csv")
```

```python
#Merging the 2 datasets through INNER JOIN
integrate = pd.merge(customers,purchases, on = "CustomerID", how = "inner")
integrate
```

```python
pd.merge(customers, purchases, left_on="CustomerID", right_on="CustomerID")
#merge() method by default performs "Inner Join"
```

```python
#This is Concatenation of Rows, but it's not a Data Integration
#This just shows how "concatenation" is different from "Data Integration using merge() with inner join"

integrate_1 = pd.concat([customers, purchases], axis=0)
integrate_1
```
