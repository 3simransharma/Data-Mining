```python
#To Control the Age Values, because they were in float, and we have to convert it into int and also fill the missing values using fillna() (Mean/Median/Mode)

#Step 1: Convert Age to numeric
cadf1["Age"] = pd.to_numeric(cadf1["Age"], errors="coerce")

#Step 2: Fill missing Age with median/mean/mode value
cadf1["Age"].fillna(cadf1["Age"].median(), inplace=True)
#We have already filled the NaN with Median here.

#Step 3: Lower Bound correct kare che
cadf1.loc[cadf1["Age"] < 1, "Age"] = 0

#Step 4: Fixing the Upper Bound
cadf1["Age"] = cadf1["Age"].clip(upper=80)

#Step 5: Converting the overall into Pure Integer (rounding-off decimal values)
cadf1["Age"] = cadf1["Age"].round().astype(int)
```

```python
#As the "FARE" column in a dataset has a very big outliers (we saw that on box plot initially in this file), we have to removing those outliers by controlling them.
#Also, some people had paid very high Fare according to their age and many have paid less, we have to controll that by IQR method.

########################################################## We are setting the fare charges according to the age.
# INFANTS (free)
cadf1.loc[cadf1["Age"] == 0, "Fare"] = 0

# CHILDREN (Age 1â€“12) realistic minimum = 10
cadf1.loc[(cadf1["Age"] >= 1) & (cadf1["Age"] <= 12) & (cadf1["Fare"] < 10), "Fare"] = 10

# TEENS (13â€“18) minimum = 12
cadf1.loc[(cadf1["Age"] >= 13) & (cadf1["Age"] <= 18) & (cadf1["Fare"] < 12), "Fare"] = 12

# ADULTS (19â€“60) minimum = 15
cadf1.loc[(cadf1["Age"] >= 19) & (cadf1["Age"] <= 60) & (cadf1["Fare"] < 15), "Fare"] = 15

# SENIORS (>60) minimum = 10
cadf1.loc[(cadf1["Age"] > 60) & (cadf1["Fare"] < 10), "Fare"] = 10



########################################################## Data Cleaning on Fare (Removing Outliers)
#Implementing IQR method

#Step 1 = Convert to numeric
cadf1["Fare"] = pd.to_numeric(cadf1["Fare"], errors="coerce")

#Step 2 = Fill missing
cadf1["Fare"].fillna(cadf1["Fare"].median(), inplace=True)

#Step 3 = Fix outliers using IQR method
Q1 = cadf1["Fare"].quantile(0.25)
Q3 = cadf1["Fare"].quantile(0.75)
IQR = Q3 - Q1
upper_limit = Q3 + 1.5 * IQR

cadf1["Fare"] = cadf1["Fare"].clip(upper=upper_limit)

#Step 4 = Round values
cadf1["Fare"] = cadf1["Fare"].round(2)
```

**IQR (Interquartile Range)** is the difference between the third quartile (Q3) and the first quartile (Q1).

It represents the spread of the middle 50% of the data and is used to detect outliers by applying the rule: values outside
[Q1 âˆ’ (1.5 Ã— IQR)] or [Q3 + (1.5 Ã— IQR)] are considered outliers.

```python
#(Method 1) Scatter plot using seaborn
colors = []

for age in cadf1["Age"]:
    if age < 1:
        colors.append("blue")
    elif age <= 12:
        colors.append("green")
    elif age <= 18:
        colors.append("orange")
    elif age <= 60:
        colors.append("red")
    else:
        colors.append("purple")

plt.figure(figsize=(8,6))
sns.scatterplot(
    x=cadf1["Age"],
    y=cadf1["Fare"],
    hue=colors,        # colours for different values
    palette="deep",
    s=50               # dot ka size
)

plt.title("Age vs Fare Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
#(Method 2) Scatter plot using Plotly
cadf1["AgeGroup"] = pd.cut(
    cadf1["Age"],
    bins=[0, 1, 12, 18, 60, 120],
    labels=["Infant", "Child", "Teen", "Adult", "Senior"]
)

fig = px.scatter(
    cadf1,
    x="Age",
    y="Fare",
    color="AgeGroup",
    title="Age vs Fare (Colored Age Groups)",
    labels={"Age": "Age", "Fare": "Fare", "AgeGroup": "Age Group"},
    size_max=10
)

fig.show()
```
