```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python
pocket_money = [20, 30, 35, 40, 40, 45, 50, 50, 55, 200,-0.00001]
```

```python
df = pd.DataFrame({"pm" : pocket_money}) #Converted from Python list into DataFrame
df  #printing DataFrame
```

```python
#Boxplot with Matplotlib and Seaborn
sns.boxplot(y=df['pm'])

plt.ylabel("Pocket Money ()")
plt.title("Box Plot of Daily Pocket Money of Students")

plt.show()
```

```python
fig = px.box(df, y='pm', title="Box Plot of Daily Pocket Money of Students")
fig.show()
```
