```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import plotly.express as px

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv("Groceries_dataset.csv")
df.head()
```

```python
basket = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).reset_index()
transactions = basket['itemDescription'].tolist()
transactions
```

```python
#Convert to One-Hot Format

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)
```

```python
#Run Apriori Algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
print("Total Frequent Itemsets:", frequent_itemsets.shape[0])
```

```python
# Generate Association Rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules = rules[rules['antecedents'].apply(lambda x: len(x) >= 1) & rules['consequents'].apply(lambda x: len(x) >= 1)]
print("Association Rules:", rules.shape[0])
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(5)
```

```python
top_items = df['itemDescription'].value_counts().head(10)

# Converting into DataFrame for Plotly
plot_df = top_items.reset_index()
plot_df.columns = ['Item', 'Count']

# Plotly bar chart
fig = px.bar(
    plot_df,
    x='Item',
    y='Count',
    title='Top 10 Most Purchased Items',
    labels={'Item': 'Item Name', 'Count': 'Purchase Count'},
    color='Count'
)

fig.show()
```

```python
import plotly.express as px

# Scatter plot of association rules
fig = px.scatter(
    rules,
    x="support",
    y="confidence",
    size="lift",
    color="lift",
    color_continuous_scale=["#a18cd1", "#fbc2eb"],  # pastel aesthetic colors
    title="Support vs Confidence (Association Rules)",
    labels={
        "support": "Support",
        "confidence": "Confidence",
        "lift": "Lift Value"
    }
)

# Improve styling
fig.update_traces(
    marker=dict(
        opacity=0.85,
        line=dict(width=1, color="white")  # bubble outline
    )
)

fig.update_layout(
    plot_bgcolor="#f7f7f7",
    paper_bgcolor="white",
    font=dict(size=14, family="Arial"),
)

fig.show()
```
