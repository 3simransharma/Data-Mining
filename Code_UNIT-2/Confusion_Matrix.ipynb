```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
```

```python
y_actual = ['COVID', 'COVID', 'COVID', 'No COVID', 'No COVID', 'No COVID']
y_predicted = ['COVID', 'COVID', 'No COVID', 'COVID', 'No COVID', 'No COVID']
```

```python
cm = confusion_matrix(y_actual, y_predicted)
cm
```

array([[2, 1], <Br> &ensp;&ensp;&ensp;&ensp;&ensp;&nbsp;
       [1, 2]])


scikit-learn arranges it as:

       [[TP  FN]
       [FP  TN]]

2 on 1st row: TP (TRUE POSITIVE) <Br>
1 on 1st row: FN (FALSE NEGATIVE) <Br>
1 on 2nd row: FP (FALSE POSITIVE) <Br>
2 on 2nd row: TN (TRUE NEGATIVE) <Br>

```python
cm_df = pd.DataFrame(
    cm,
    index=['Actual COVID', 'Actual No COVID'],
    columns=['Predicted COVID', 'Predicted No COVID']
)

cm_df
```

```python
accuracy = accuracy_score(y_actual, y_predicted)
print("Accuracy:", accuracy)
```

**pos_label:** We choose one class as the positive class (the class we care most about).

```python
precision = precision_score(y_actual, y_predicted, pos_label='COVID')
print("precision:", precision)
```

```python
recall = recall_score(y_actual, y_predicted, pos_label='COVID')
print("Recall:", recall)
```

```python
f1 = f1_score(y_actual, y_predicted, pos_label='COVID')
print("F1 Score:", f1)
```

```python
print(classification_report(y_actual,y_predicted))
```

```python
sb.heatmap(cm, annot=True, fmt='d',
            xticklabels=['COVID', 'No COVID'],
            yticklabels=['COVID', 'No COVID'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
